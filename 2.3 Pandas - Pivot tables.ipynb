```python
import pandas as pd
import numpy as np
link = "https://raw.githubusercontent.com/WildCodeSchool/data-training-resources/main/quests/weather2019.csv"
df_weather = pd.read_csv(link)
```

```python
df_weather.head()
```

PIVOT TABLE

There are 2 syntaxes, choose your preferred :

As a method : YOUR_DATAFRAME.pivot_table()
As a function : pd.crosstab()

```python
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

```python
# Other syntax:
pd.crosstab(values = df_weather['MAX_TEMPERATURE_C'],
            index = df_weather['MONTH'], columns ="",
            aggfunc = 'mean')
```

Create a Pivot table with :

- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```python
df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'DAY', columns ='MONTH')
```

Pivot table with :

- Months on rows
- the sum of Precipitations on values

Then display it as a column chart. (optionnal)

```python
df_precipitation = df_weather.pivot_table(values = 'PRECIP_TOTAL_DAY_MM', index = 'MONTH',  aggfunc = 'sum')
print(df_precipitation)
#df_precipitation.plot(kind='bar', figsize=(15,5), rot=0, title = 'Precipitation en mm')
```

Pivot table with :

- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```python
df_weather.pivot_table(index=["MONTH"],values=["MAX_TEMPERATURE_C","SUNHOUR"],
               aggfunc={"MAX_TEMPERATURE_C":np.mean,"SUNHOUR":np.mean},fill_value=0)
```

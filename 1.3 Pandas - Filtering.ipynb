# Mission 4

Display only the rows with films that belong to the lowest rated genre (based on the average rating of the films in this genre)
[This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.mean.html)

```python
# Average rating per genre
df.groupby(["genre"])["star_rating"].mean().round(2)

# Other possible solution :
#df.groupby('genre').mean().sort_values('star_rating')
```

```python
# Genre with minimum average rating
df.groupby(["genre"])["star_rating"].mean().idxmin()
```

```python
# Display only the rows with films that belong to the lowest rated genre
genre_rating = df.groupby(["genre"])["star_rating"].mean().round(2)
df2 = df.merge(genre_rating, on='genre',suffixes=('', '_avg'))
df2[
    df2['star_rating_avg']==df2['star_rating_avg'].min()
    ]

# Other possible solution :
# genre_low = df.groupby('genre').star_rating.mean()
# df[df['genre'] == genre_low.index[genre_low.argmin()]]

# Other possible solution :
df[df.genre.isin((df.groupby("genre").star_rating.mean().nsmallest(1)).index)]
```

# Mission 5

Display films which are longer than 200 minutes **and** which belong to a genre with a name containing the letter 'h' (either lower case or upper case).

You should get only one film: `Malcolm X`


```python
# Films longer than 200 min
df.loc[(df['duration'] > 200)]
```

```python
# genre with a name containing the letter 'h' (lower or upper)
df.loc[df['genre'].astype(str).str.contains("h|H")]
```

```python
# Films longer than 200 min AND genre with a name containing the letter 'h' (lower or upper)
df.loc[((df['duration'] > 200)) & (df['genre'].astype(str).str.contains("h|H"))]
```

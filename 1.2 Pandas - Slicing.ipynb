# Mission 5

Rename the columns [of this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv), try using the following 2 different methods to achieve this:
- Setting column names directly when reading the file
- Editing after creating the dataframe ([This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html))

```python
import pandas as pd

# Reading the dataset into a DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv')

# Method 1: Setting column names directly when reading the file
new_column_names = ['city', 'colors_reported', 'shape_reported', 'state', 'time']
df_direct = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', names=new_column_names, header=0)

print("Method 1: Setting column names directly when reading the file")
print(df_direct.head())

# Method 2: Editing column names after creating the DataFrame
df_renamed = df.rename(columns={
    'City': 'city',
    'Colors Reported': 'colors_reported',
    'Shape Reported': 'shape_reported',
    'State': 'state',
    'Time': 'time'
})

print("\nMethod 2: Editing column names after creating the DataFrame")
print(df_renamed.head())
```
